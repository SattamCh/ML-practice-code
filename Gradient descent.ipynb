```python
import numpy as np
import matplotlib.pyplot as plt


x = np.array([1, 2, 3, 4, 5], dtype=float)
y = np.array([5, 6, 10, 13, 11], dtype=float)
#given parameters
b0 = 3.3
b1 = 1.9
learning_rate = 1.1
iterations = 10


def cost_function(x, y, b0, b1):
    n = len(x)
    y_pred = b0 + b1 * x
    return (1 / (2 * n)) * np.sum((y_pred - y) ** 2)

# -----------------------------
# Gradient Descent
# -----------------------------
cost_history = []
n = len(x)

print("Iteration |     b0     |     b1     |     Cost")
print("------------------------------------------------")

for i in range(iterations):
    y_pred = b0 + b1 * x

    # Gradients
    db0 = (1 / n) * np.sum(y_pred - y)
    db1 = (1 / n) * np.sum((y_pred - y) * x)

    # Update parameters
    b0 = b0 - learning_rate * db0
    b1 = b1 - learning_rate * db1

    cost = cost_function(x, y, b0, b1)
    cost_history.append(cost)

    print(f"{i+1:^9} | {b0:9.4f} | {b1:9.4f} | {cost:9.4f}")

# -----------------------------
# Cost curve
# -----------------------------
plt.figure()
plt.plot(range(1, iterations + 1), cost_history, marker='o')
plt.xlabel("Iterations")
plt.ylabel("Cost (MSE)")
plt.title("Gradient Descent Learning Curve")
plt.show()
```
